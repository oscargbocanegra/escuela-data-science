```python
# Creando un path con la ruta de los aarchivos para cargar los RDD.
path = "/home/jovyan/work/files/"

#Creando el primer RDD usando textfile para cargar los datos del archivo.
equiposOlimpicosRDD = sc.textFile(path+"paises.csv") \
    .map(lambda line : line.split(","))
```

```python
#opcion take para mostrar la cantidad de lineas.
equiposOlimpicosRDD.take(15)
```

```python
# Realizando opoeraciones con los RDD.
# Cuantos paises participan en los juegos olimpicos, partiendo del campo 3 sigla.
# distinct() para validar los duplicados.
# count() para contar los valores de la lista.
equiposOlimpicosRDD.map(lambda x : (x[2])).distinct().count()
```

```python
# Cuantos registros existen en el dataframe
equiposOlimpicosRDD.count()
```

```python
# Operaciones Group by Cuantos equipo posee a cada equipo.
equiposOlimpicosRDD.map(lambda x: (x[2],x[1])).groupByKey().mapValues(len).take(15)
```

```python
# Operaciones Group by Cuantos equipo posee a cada equipo mostrado en formato lista.
equiposOlimpicosRDD.map(lambda x: (x[2],x[1])).groupByKey().mapValues(list).take(15)
```

```python
# Operaciones Group by Cuantos equipo posee a cada equipo mostrado en formato lista solo para los equipos argentinos.
equiposArgentinos = equiposOlimpicosRDD.filter(lambda l: "ARG" in l)
equiposArgentinos.collect()
```

```python
# Esta es una opcion parra obtener el conteo de la cantidad de documentos en el arhcivo cargado
equiposOlimpicosRDD.count()
```

```python
# Esta es una opcion parra obtener el conteo aproximado hasta el limite de tiempo en milisegundos que se coloca en countApprox(20)
equiposOlimpicosRDD.countApprox(20)
```

```python
# Cargando archivos.
# Creando RDD para deportista olimpico con el archivo deportista.csv
deportistaOlimpicoRDD = sc.textFile(path+"deportista.csv").map(lambda l : l.split(","))
# Creando RDD2 para deportista olimpico con el archivo deportista2.csv
deportistaOlimpicoRDD2 = sc.textFile(path+"deportista2.csv").map(lambda l : l.split(","))
```

```python
deportistaOlimpicoRDD.collect()
```

```python
# En esta seccion se realiza la union de los 2 Dataframe de deportista. con UNION esta operación se encarga de los duplicados.
deportistaOlimpicoRDD = deportistaOlimpicoRDD.union(deportistaOlimpicoRDD2)
```

```python
# Para hacer los JOIN se seleccionan primero los id
deportistaOlimpicoRDD.map(lambda l: [l[-1],l[:-1]]).join(equiposOlimpicosRDD.map(lambda x : [x[0],x[2]])).take(6)
```

```python
# Para hacer los JOIN se seleccionan primero los id en este caso se usa takeSample para escoger una muestra aleatoria
deportistaOlimpicoRDD.map(lambda l: [l[-1],l[:-1]]).join(equiposOlimpicosRDD.map(lambda x : [x[0],x[2]])).takeSample(False,6,25)
```

```python
# Cargando el archivo de resultados.
resultado = sc.textFile(path+"resultados.csv").map(lambda l : l.split(","))
```

```python
# Filtrar por el resultado de los jugadores que no han obtenido una medalla.
resultadoGanador = resultado.filter(lambda l : 'NA' not in l[1])
```

```python
# Visualizando el contenido de resultado Ganador
resultadoGanador.take(10)
```

```python
# Realizando la union de deportistas y equipos olimpicos para optener el dato de deportistas por paises.
deportistaPaises = deportistaOlimpicoRDD.map(lambda l : [l[-1], l[:-1]]).join(equiposOlimpicosRDD.map(lambda x :[x[0], x[2]]))
```

```python
deportistaPaises.join(resultadoGanador).take(6)
```

```python
# Operaciones Numericas.
# Ejercicio - Cual es el valor de puntaje que tienen los paises a lo largo del tiempo.
valoresMedallas = {'Gold' : 7, 'Silver' : 5, 'Bronze' : 4}
```

```python
# Guardando el valor de los join en valriable paisesMedallas para usarlo mas adelante
paisesMedallas = deportistaPaises.join(resultadoGanador)
```

```python
# Realizando Extracción para obtener los 2 ultimos valores de medalla y pais.
paisesMedallas = paisesMedallas.map(lambda x: (x[1][0][-1], valoresMedallas[x[1][1]]))
```

```python
# Realizando las sumas importando la libreria add para realizar las sumas.
from operator import add
# Realizando una reduccion por llave para indicar que se hace una suma
conclusion = paisesMedallas.reduceByKey(add).sortBy(lambda x :x[1],ascending=False)
```

```python
# Puntaje historico que poseen los paises que jugaron en las olimpiadas.
conclusion.take(10)
```
