```python
# Creando un path con la ruta de los aarchivos para cargar los RDD.
path = "/home/jovyan/work/curso-apache-spark-platzi/files/"

#Creando el primer RDD usando textfile para cargar los datos del archivo.
equiposOlimpicosRDD = sc.textFile(path+"paises.csv") \
    .map(lambda line : line.split(","))
```

```python
#opcion take para mostrar la cantidad de lineas.
equiposOlimpicosRDD.take(15)
```

```python
# Realizando opoeraciones con los RDD.
# Cuantos paises participan en los juegos olimpicos, partiendo del campo 3 sigla.
equiposOlimpicosRDD.map(lambda x : (x[2])).distinct().count()
```

```python
# Cuantos registros existen en el dataframe
equiposOlimpicosRDD.count()
```

```python
# Operaciones Group by Cuantos equipo posee a cada equipo.
equiposOlimpicosRDD.map(lambda x: (x[2],x[1])).groupByKey().mapValues(len).take(15)
```

```python
# Operaciones Group by Cuantos equipo posee a cada equipo mostrado en formato lista.
equiposOlimpicosRDD.map(lambda x: (x[2],x[1])).groupByKey().mapValues(list).take(15)
```

```python
# Operaciones Group by Cuantos equipo posee a cada equipo mostrado en formato lista solo para los equipos argentinos.
equiposArgentinos = equiposOlimpicosRDD.map(lambda l: "ARG" IN L)
equiposArgentinos.
```
